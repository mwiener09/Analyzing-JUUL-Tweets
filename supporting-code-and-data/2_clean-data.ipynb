````python
#create data frame from tweets
df = pd.DataFrame()
for json_file in json_file_list:
    tweet_data = [] # clear tweet data
    with open(json_file, 'r') as f:
        print(json_file)
        for line in f.readlines():
            tweet_data.append(json.loads(line))
    # create local df with all of the file's info
    local_df = pd.DataFrame() # clear local df
    #general info
    local_df['id'] = list(map(lambda tweet: tweet['id_str'], tweet_data))
    local_df['date_scraped'] = json_file[len(json_file)-15:len(json_file)-5].replace('-','') #last date
    local_df['date_scraped'] = local_df.date_scraped.astype('int64') # as number
    local_df['tweet_date'] = list(map(lambda tweet: tweet['created_at'], tweet_data))
    # format datetime
    local_df['tweet_datetime'] = pd.to_datetime(local_df.tweet_date)
    local_df['tweet_date'] = local_df.tweet_datetime.dt.date
    local_df['tweet_hour'] = local_df.tweet_datetime.dt.hour
    local_df['tweet_time'] = local_df.tweet_datetime.dt.hour.apply(lambda time: 'nighttime' if (time <= 4) | (time >= 17) else 'daytime') # 5pm to 4am
    local_df['tweet_day_of_week'] = local_df.tweet_datetime.dt.weekday_name
    local_df['text'] = list(map(lambda tweet: tweet['text'], tweet_data))
    local_df['lang'] = list(map(lambda tweet: tweet['lang'], tweet_data))

    # user information
    local_df['screen_name'] = list(map(lambda tweet: tweet['user']['screen_name'], tweet_data))
    local_df['location'] = list(map(lambda tweet: tweet['user']['location'], tweet_data))
    local_df['protected'] = list(map(lambda tweet: tweet['user']['protected'], tweet_data))
    local_df['profile_description'] = list(map(lambda tweet: tweet['user']['description'], tweet_data))
    local_df['followers_count'] = list(map(lambda tweet: tweet['user']['followers_count'], tweet_data))
    local_df['friends_count'] = list(map(lambda tweet: tweet['user']['friends_count'], tweet_data))
    local_df['listed_count'] = list(map(lambda tweet: tweet['user']['listed_count'], tweet_data))
    local_df['favourites_count'] = list(map(lambda tweet: tweet['user']['favourites_count'], tweet_data))
    local_df['geo_enabled'] = list(map(lambda tweet: tweet['user']['geo_enabled'], tweet_data))
    local_df['verified'] = list(map(lambda tweet: tweet['user']['verified'], tweet_data))
    local_df['statuses_count'] = list(map(lambda tweet: tweet['user']['statuses_count'], tweet_data))
    # tweet info
    local_df['retweet_count'] = list(map(lambda tweet: tweet['retweet_count'], tweet_data))
    local_df['favorite_count'] = list(map(lambda tweet: tweet['favorite_count'], tweet_data))
    # tweet location
    local_df['longitude'] = list(map(lambda tweet: tweet['coordinates']['coordinates'][0] 
                                      if tweet['coordinates'] != None else 'NaN', tweet_data))
    local_df['latitude'] = list(map(lambda tweet: tweet['coordinates']['coordinates'][1] 
                                      if tweet['coordinates'] != None else 'NaN', tweet_data))
    if (df.shape[0] == 0):
        df = local_df
    else:
        df = pd.concat([df, local_df],ignore_index = True)
    print(df.shape)
````

````python
# drop duplicates
print(df.shape)
df.drop_duplicates(inplace = True) # drop duplicates
print(df.shape)
````

````python
# because of the way I scraped tweets, there is some duplication
# the following cells address this issue

# create an id based on tweet text and datetime
# join this id to original dataset
id_lookup = df[['tweet_datetime','text']].drop_duplicates() #,'screen_name'
id_lookup['unique_tweet_id'] = list(range(1,id_lookup.shape[0]+1))
df = pd.merge(df, id_lookup, how='left', left_on=['tweet_datetime','text'], 
          right_on=['tweet_datetime','text'])


# create an short tweet_id based on original id
# join this id to original dataset
short_id = df[['id']].drop_duplicates()
short_id.sort_values(by = 'id', inplace = True) # preserve order
short_id['short_tweet_id']= list(range(1,short_id.shape[0]+1))
df = pd.merge(df, short_id, how='left', left_on=['id'], 
          right_on=['id'])
````

````python
# we want 1 row for each unique_tweet_id
df.head()
df[df.unique_tweet_id == 1] # example of dupe
````

````python
# most duplicated tweets are pulled again on a different day 
# some cases of same day, but different counts of friends, followers, etc
# make a numeric hash column to be used to order tweets by unique_tweet_id
# and remove duplicates
df['recent_stat_summary'] = df.statuses_count + df.retweet_count + df.favorite_count + df.followers_count +\
                            df.friends_count + df.listed_count + df.favourites_count
#df['recent_stat_summary'] = [str(n)[-2:] for n in df['recent_stat_summary']] # just take last two digits
# priority - date scraped, then recenet stat changes
df['find_most_recent'] =  df.date_scraped.astype('str') + df.recent_stat_summary.astype('str') #+\
                            #df.short_tweet_id.astype('str')
df['find_most_recent'] = df.find_most_recent.astype(float)
````

````python
# make sure things look as expected
print(df.shape)
df.dtypes
df.head()
````

````python
# create master list of rows to keep using the most recent 
# numeric index 'find most recent'
series_lookup = df.groupby('unique_tweet_id').find_most_recent.max()

# create lookup table for max index (most recent)
keep_id = pd.DataFrame()
keep_id['unique_tweet_id'] = series_lookup.index
keep_id['most_recent_id'] = series_lookup.values

# join the most recent id back on original df
df = pd.merge(df, keep_id, how='left', left_on='unique_tweet_id', right_on='unique_tweet_id')
df.head(100)
````

````python
# only keep most recently scraped tweet and check unique values
# should roughly match
print(df.shape)
df = df[df.most_recent_id == df.find_most_recent].drop_duplicates()
print(df.shape)
print(len(df.unique_tweet_id.unique()))
````

````python
# filter out non-juul text - if word is found in screen name, profile 
# description rather than tweet text
juul = df[df.text.str.lower().str.contains('juul')] # if description, screen name 
no_juul = df[~df.text.str.lower().str.contains('juul')]

# get rid of retweets
sum(juul.text.str[0:2] == 'RT')
juul = juul[juul.text.str[0:2] != 'RT']
print(juul.shape)
len(juul.unique_tweet_id.unique())
````
